### Answer to Question 1

First of all, let's nondimensionalize the equation:

\begin{align}
        a_1 x^9 - a_2 x^3 - a_3 = 0
\end{align}

Let's write $x$ as a function of a new unknown $y$:

\begin{align}
        x = \left(\frac{a_3}{a_2}\right)^\frac{1}{3}y
\end{align}

and let's plug in into the equation:

\begin{align}
        a_1 \left( \frac{a_3}{a_2} \right)^3 y^9 - a_2 \left( \frac{a_3}{a_2} \right) y^3 - a_3 = 0
\end{align}

Now, let's divide both sides by $a_3$:

\begin{align}
        a_1 \frac{a_3^2}{a_2^3} y^9 - y^3 - 1 = 0
\end{align}

Renaiming $y$ as $x$:

\begin{align}
        a_1 \frac{a_3^2}{a_2^3} x^9 - x^3 - 1 = 0
\end{align}

where we can indicate the factor located in front of $x^9$ as $\epsilon$:

\begin{align}
        \epsilon = a_1 \frac{(a_3)^2}{(a_2)^3}
\end{align}

The equation could then be rewritten as:

\begin{align}
        \epsilon x^9 - x^3 + 1 = 0
\end{align}

### Answer to Question 3

As requested, we expand around the root $x = 1$. The expansion can be written as :

\begin{align}
  x = 1 +\sum_{n=1}^{\infty} a_n \epsilon^n.
\end{align}

At this point, the expansion can be plugged in into the ninth degree polynomial, $\epsilon x^9 - x^3 + 1 = 0$:

\begin{align}
  \epsilon \left( 1 +\sum_{n=1}^{\infty} a_n \epsilon^n \right)^9 - \left( 1 +\sum_{n=1}^{\infty} a_n \epsilon^n \right)^3 + 1 = 0.
\end{align}

Below, we explicitly expand each term and we group the terms according to the power of $\epsilon$ (namely, $\epsilon$, $\epsilon^2$, etc...). 

```python
epsilon, a_1, a_2, a_3 = symbols('epsilon a_1 a_2 a_3')
expansion = expand(epsilon*(1+a_1*epsilon+a_2*epsilon**2+a_3*epsilon**3)**9-(1+a_1*epsilon+a_2*epsilon**2+a_3*epsilon**3)**3+1)
collect(expansion,epsilon)
```

In order to calculate the radius of convergence, we can compute $a_1$, $a_2$ and $a_3$ from the last three terms:

\begin{align}
  1-3a_1 = 0,
\end{align}

which implies:

\begin{align}
  a_1 = \frac{1}{3}
\end{align}

From the term involving the second power of $\epsilon$, we get:

\begin{align}
  -3a_1^2 + 9a_1 - 3a_2 = 0.
\end{align}

Plugging into it the solution for $a_1$, we easily get:

\begin{align}
  -3\left( \frac{1}{3} \right)^2 + 9\left( \frac{1}{3} \right) - 3a_2 = 0, 
\end{align}

which gives:

\begin{align}
  a_2 = \frac{8}{9}.
\end{align}

Now, we can plug in both $a_1$ and $a_2$ into $-a_1^3 + 36a_1^2 - 6a_1a_2 + 9a_2 - 3a_3$  in order to get $a_3$:

\begin{align}
  a_3 = \frac{1}{3}(-a_1^3 + 36a_1^2 - 6a_1a_2 + 9a_2).
\end{align}

```python
a_1 = 1/3
a_2 = 8/9
a_3 = (1/3)*(-a_1**3+36*a_1**2-6*a_1*a_2+9*a_2)
print(a_3)
```

```python
epsilon = abs(a_2/a_3)
print(epsilon)
```

We estimate the radius of convergence to be:

\begin{align}
  \epsilon = \left| \frac{a_2}{a_3} \right| \approx 0.262
\end{align}

### Answer to Question 5

In order to analytically find the radius of convergence, a system of two equation involving $f(x) = 0$ and $\frac{df}{dx} = 0$ has to be solved.

Therefore, we have:

\begin{align}
  \begin{cases} \epsilon x^9 - x^3 + 1 = 0\\
9\epsilon x^8 -3x^2 = 0\\
  \end{cases}
\end{align}

From the second equation, we get:

\begin{align}
  x = \frac{1}{(3\epsilon)^\frac{1}{6}}
\end{align}

Plugging it into the first one, we get:

\begin{align}
  \epsilon\left( \frac{1}{(3\epsilon)^\frac{1}{6}} \right)^9 - \left( \frac{1}{(3\epsilon)^\frac{1}{6}} \right)^3 + 1 = 0
\end{align}

which implies:

\begin{align}
  \frac{1}{3^\frac{3}{2}}\epsilon^{-\frac{1}{2}} - \frac{1}{3^\frac{1}{2}}\epsilon^{-\frac{1}{2}} + 1 = 0
\end{align}

Then:

\begin{align}
  \epsilon = \left( \frac{1}{3^\frac{1}{2}} - \frac{1}{3^\frac{3}{2}} \right)^2
\end{align}

```python
epsilon = ((1/3)**(1/2)-(1/3)**(3/2))**2
print(epsilon)
```

In conclusion:

\begin{align}

* Mean free path of an "air molecule", $\bar l$ [$L$]
* Mean speed of an "air molecule", $\bar v$ [$L/T$]  

The quantity we are looking for is:  
* Time, $t$ [$T$]

A generic air particle escapes the paper cup by random-walking. Let's consider an air particle located at the center of the paper cup and let's assume it can only go up by random-walking. The mean-square path can be expressed by $2\bar l^2$, which implies the mean $z$-component is $2\bar l^2/3$. The number of steps in the random walk can be calculated by $\bar v/\bar l$.

The time $t$ can be calculated as:

\begin{align}
  t \sim \frac{\bar D^2}{\frac{\bar v}{\bar l}\cdot \frac{2\bar l^2}{3}}
\end{align}

where $\bar D^2$ is the mean-square displacement.

We assume:

* $\bar D^2 \approx 2.5 \cdot 10^{-3}$ $m^2$, 
* $\bar l \approx 10^{-7}$ $m$, 
* $\bar v \approx 5\cdot 10^2$ $m/s$, 

which gives:

\begin{align}
t \approx 75 s
\end{align}

which is an underestimate, since we assumed that the air particle could only move upward while random-walking.

### Answer to Question 2

The quantities we know and their units are:

* The radius of the Sun, $R_{S}$ [$L$]
* The radius of the Moon, $R_{M}$ [$L$]
* The distance between the Earth and the Sun, $d_{E-S}$ [$L$]
* The distance between the Earth and the Moon, $d_{E-M}$ [$L$]

The quantity we aim to find is:

* The ratio between the average density of the Moon to that of the Sun, $\rho_M/\rho_S$, which is a nondimensional quantity. 

The gravitational acceleration of the ocean due to the Moon is:

\begin{align}
  a_{O-M} = \frac{G\cdot M_M}{d_{E-M}^2}
\end{align}

where $G$ is the universal constant of gravitation. 

Similarly, the gravitational acceleration of the ocean due to the Sun is:

\begin{align}
  a_{O-S} = \frac{G\cdot M_S}{d_{E-S}^2}
\end{align}

We know that the tidal effect is proportional to the gradient of the gravitational field, $M/d^3$.

\begin{align}
  \frac{M_M\cdot d_{E-S}^3}{M_S\cdot d_{E-M}^3} = \frac{7}{3}
\end{align}

We can then write:

\begin{align}
  \frac{M_M}{M_S} = \frac{7}{3}\cdot \frac{d_{E-M}^3}{d_{E-S}^3}
\end{align}


Multiplying and dividing the left-hand side by $V_M$ and $V_S$, the volume of the Moon and the Sun, respectively, we get:

\begin{align}
  \frac{\rho_M}{\rho_S}\cdot \frac{V_M}{V_S} = \frac{7}{3}\cdot \frac{d_{E-M}^3}{d_{E-S}^3}
\end{align}

which can be rewritten as:

\begin{align}
  \frac{\rho_M}{\rho_S} = \frac{7}{3}\cdot \frac{d_{E-M}^3}{d_{E-S}^3}\frac{V_S}{V_M} = \frac{7}{3}\cdot \frac{d_{E-M}^3}{d_{E-S}^3}\frac{\frac{4}{3}R_S^3}{\frac{4}{3}R_M^3} = \frac{7}{3}\cdot \frac{d_{E-M}^3}{d_{E-S}^3}\frac{R_S^3}{R_M^3}
\end{align}

Assuming:

* $d_{E-M} \approx 3.9 \cdot10^8$,
* $d_{E-S} \approx 1.5\cdot10^{11}$,
* $R_M \approx 1.7\cdot10^6$,
* $R_S \approx 7.0\cdot10^8$,

we get:

\begin{align}
 \frac{\rho_M}{\rho_S} \approx 2.86
\end{align}

* The angle of deflection, $\theta$ dimensionless

We have $n = 4$ dimensional variables and $k = 3$ independent dimensions. By Buckingham's theorem, the number of independent $\Pi$-groups is $p = n-k = 1$.

$$ \Pi_1 = \frac{GM}{rc^2}$$.

Note that the angle of deflection $\theta$ is dimensionless and constitutes the second independent $\Pi$ group, $\Pi_2 = \theta$.

From Buckingham's theorem:

$$ f(\Pi_1, \Pi_2) = 0 $$

which implies:

$$ \theta = f(\frac{GM}{rc^2}) $$

A very simple scaling for $\theta$ could be:

$$ \theta \sim \frac{GM}{rc^2} $$.

Assuming:

* $G \approx 6.7\cdot 10^{-11} m^3kg^{-1}s^{-2}$
* $M = 2.0\cdot 10^{30}$ (mass of the Sun)
* $r = 7.0\cdot 10^8 m$ (radius of the Sun)
* $c = 3.0\cdot 10^{8} ms^{-1}$

we get:

$$ \theta \approx 2.1\cdot 10^{-6}rad$$
